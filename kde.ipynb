```python
# import scipy kde
from scipy.stats import gaussian_kde
```

```python
import numpy as np
```

```python
x = np.random.normal(0, 1, 1_000_000)
```

```python
f = gaussian_kde(x)
```

```python
%%timeit
f(0)
```

```python
xs = np.linspace(-5, 5, 100)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%%timeit
y = f(xs)
```

```python
plt.plot(xs, y)
```
